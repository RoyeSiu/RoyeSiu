```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

print(f'Tensorflow version is {tf.__version__}')
print(f'Tensorflow dataset version is {tfds.__version__}')
```

```python
(raw_train, raw_validation, raw_test), metadata = tfds.load(
    'cats_vs_dogs',
    split=['train[:85%]','train[85%:90%]','train[90%:]'],
    with_info=True,
    as_supervised=True
)
```

```python
for image,label in raw_train.take(3):
  plt.figure()
  plt.imshow(image)
  plt.title('Image')
```

```python
image,label = next(iter(raw_train.take(1)))
plt.figure()
plt.imshow(image)
plt.title('Image')
IMG_SIZE = 160
def resize_image(image,label):
  image = tf.cast(image,tf.float32)
  image = (image/127.5)-1
  image = tf.image.resize(image,(IMG_SIZE,IMG_SIZE))
  return image,label
# print(format_example(image,label))
image,label = resize_image(image,label)
plt.figure()
plt.imshow(image)
plt.title('Image')
```
